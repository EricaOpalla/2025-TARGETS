```python
function onEdit(e) {
  var sheet = e.source.getActiveSheet();
  var ncbcSheet = e.source.getSheetByName("NCBC");
  var databaseSheet = e.source.getSheetByName("Database");

  if (!ncbcSheet || !databaseSheet) return;

  // If dropdown (H1) is changed, clear NCBC inputs and load data from Database
  if (sheet.getName() === "NCBC" && e.range.getA1Notation() === "H1") {
    var selectedValue = e.range.getValue(); // Get selected dropdown value
    if (selectedValue) {
      ncbcSheet.getRange("H4:AE31").clearContent(); // Clear previous data
      retrieveDataFromDatabase(selectedValue, ncbcSheet, databaseSheet); // Load new data
    }
  }

  // If editing an input in NCBC (H4:AE31), store data in Database
  else if (sheet.getName() === "NCBC" && e.range.getRow() >= 4 && e.range.getRow() <= 31 && e.range.getColumn() >= 8 && e.range.getColumn() <= 31) {
    var rowMapping = {
      4: 45, 5: 46, 6: 51, 7: 41, 8: 42, 9: 43, 10: 44, 11: 45,
      12: 46, 13: 47, 14: 48, 15: 49, 16: 50, 17: 51, 18: 52, 19: 53, 
      20: 54, 21: 55, 22: 56, 23: 57, 24: 58, 25: 59, 26: 60, 27: 61,
      28: 62, 29: 63, 30: 64, 31: 65
    };

    var selectedValue = ncbcSheet.getRange("H1").getValue();
    if (selectedValue && rowMapping[e.range.getRow()]) {
      var databaseStorage = getDatabaseStorage(selectedValue);

      if (databaseStorage) {
        var ncbcRow = e.range.getRow();
        var dbRow = rowMapping[ncbcRow];

        var ncbcRowData = ncbcSheet.getRange(ncbcRow, 8, 1, 24).getValues();
        var dbRange = databaseSheet.getRange(dbRow, getColumnIndex(databaseStorage), 1, 24);
        dbRange.setValues(ncbcRowData);
      }
    }
  }
}

// Function to map dropdown values to storage locations in Database
function getDatabaseStorage(selectedValue) {
  var databaseMapping = {
    "NC PANGLAO": "C", "NC TUBIGON": "AB", "NC LOON": "BA",
    "NC ANTEQUERA": "BZ", "NC CALAPE": "CY", "NC BOSS": "DX",
    "NC CORELLA": "EW", "NC SIKATUNA": "FV", "NC CATIGBIAN": "GU",
    "NC INABANGA": "HT", "NC SAGBAYAN": "IS", "NC BUENAVISTA": "JR",
    "NC GETAFE": "KQ", "NC TALIBON": "LP", "NC TRINIDAD": "MO",
    "NC SAN MIGUEL": "NN", "NC UBAY": "OM", "NC MABINI": "PL",
    "NC CARMEN": "QK", "NC SIERRA BULLONES": "RJ", "NC JAGNA": "SI",
    "NC GUINDULMAN": "TH", "NC ANDA": "UG", "NC LOAY": "VF",
    "NC DIMIAO": "WE", "NC VALENCIA": "XD", "NC LOBOC": "YC",
    "NC SEVILLA": "ZB", "NC BATUAN": "AAA", "NC BPO": "AAZ"
  };

  return databaseMapping[selectedValue] || null;
}

// Function to retrieve the correct database column index
function getColumnIndex(columnLetter) {
  var columnIndex = 0;
  for (var i = 0; i < columnLetter.length; i++) {
    columnIndex = columnIndex * 26 + (columnLetter.charCodeAt(i) - 64);
  }
  return columnIndex;
}

// Function to retrieve data from Database and populate NCBC
function retrieveDataFromDatabase(selectedValue, ncbcSheet, databaseSheet) {
  var rowMapping = {
    4: 45, 5: 46, 6: 51, 7: 41, 8: 42, 9: 43, 10: 44, 11: 45,
    12: 46, 13: 47, 14: 48, 15: 49, 16: 50, 17: 51, 18: 52, 19: 53, 
    20: 54, 21: 55, 22: 56, 23: 57, 24: 58, 25: 59, 26: 60, 27: 61,
    28: 62, 29: 63, 30: 64, 31: 65
  };

  var databaseStorage = getDatabaseStorage(selectedValue);
  if (databaseStorage) {
    for (var ncbcRow in rowMapping) {
      var dbRow = rowMapping[ncbcRow];

      var dbRange = databaseSheet.getRange(dbRow, getColumnIndex(databaseStorage), 1, 24);
      var storedValues = dbRange.getValues();

      var inputCell = ncbcSheet.getRange(parseInt(ncbcRow), 8, 1, 24);
      inputCell.setValues(storedValues);
    }
  }
}
```

```python
function onEdit(e) {
  var sheet = SpreadsheetApp.getActiveSpreadsheet().getSheetByName("REGULAR STAFF");
  var dbSheet = SpreadsheetApp.getActiveSpreadsheet().getSheetByName("DATABASE");
  var editedCell = e.range;
  
  // Mapping of dropdown values to database columns
  var columnMap = {
    "REGIS": "AJV",
    "BALABA": "AJW",
    "ISRAEL": "AJX",
    "PANONG": "AJY",
    "TAY": "AJZ",
    "ESCOBILLA": "AKA",
    "REQUILLO": "AKB"
  };
  
  // Check if the edit happened in E1
  if (editedCell.getA1Notation() === "E1") {
    var filterValue = editedCell.getValue(); // Get selected value from dropdown
    var targetRange = sheet.getRange("E4:E87");
    targetRange.clearContent(); // Clear existing data
    
    if (filterValue in columnMap) {
      var columnLetter = columnMap[filterValue];
      var dbRange = dbSheet.getRange(columnLetter + "4:" + columnLetter + "87");
      var data = dbRange.getValues();
      targetRange.setValues(data); // Populate E4:E87 with filtered data
    }
  }
  
  // Automatically transfer data from REGULAR STAFF to DATABASE
  if (editedCell.getRow() >= 4 && editedCell.getRow() <= 87 && editedCell.getColumn() == 5) { // Column E
    var selectedValue = sheet.getRange("E1").getValue();
    if (selectedValue in columnMap) {
      var columnLetter = columnMap[selectedValue];
      var dbCell = dbSheet.getRange(columnLetter + editedCell.getRow());
      dbCell.setValue(editedCell.getValue());
    }
  }
}
```

```python
function onEdit(e) {
  var sheet = e.source.getSheetByName('REGULAR STAFF');
  var dbSheet = e.source.getSheetByName('DATABASE');
  
  var dropdownRange = sheet.getRange('H1:AE1').getValues()[0]; // Get dropdown values
  var inputRange = sheet.getRange('H4:AE87'); // Input data range (for storing & retrieving)

  var storageMapping = {
    "REGIS": "ABY4:ACV87",
    "BALABA": "ACX4:ADU87",
    "ISRAEL": "ADW4:AET87",
    "PANONG": "AEV4:AFS87",
    "TAY": "AFU4:AGR87",
    "ESCOBILLA": "AGT4:AHQ87",
    "REQUILLO": "AHS4:AIP87"
  };

  var editedCell = e.range;

  // Step 1: If a dropdown value changes, clear the input range (H4:AE87) and retrieve stored data
  if (editedCell.getRow() === 1 && editedCell.getColumn() >= 8 && editedCell.getColumn() <= 31) {
    sheet.getRange('H4:AE87').clearContent(); // Clear the input range

    var selectedValue = editedCell.getValue(); // Get new dropdown value
    var storageRange = storageMapping[selectedValue]; // Get corresponding database range

    if (storageRange) {
      var storedData = dbSheet.getRange(storageRange).getValues(); // Get stored data
      inputRange.setValues(storedData); // Return data back to the dashboard
    }
  }

  // Step 2: Store new input data in the corresponding database range
  for (var i = 0; i < dropdownRange.length; i++) {
    var storageRange = storageMapping[dropdownRange[i]];
    
    if (storageRange) {
      var targetRange = dbSheet.getRange(storageRange);
      targetRange.setValues(inputRange.getValues()); // Store input data
    }
  }
}
```
